# **Enhancing Visualizations**
### _**Adding Fills to Log Plots**_


Filling the log plots not only nehances the visualisations but also help petrophysists to better understand the data and help them identify regions wit maximum and minimum values.Log plots are very important in exploring and visualizing well logs data.

## Importing Modules and Loading Data

```python
import pandas as pd
import lasio
import matplotlib.pyplot as plt
import numpy as np
```

```python
las = lasio.read('../Data/inputs/16_7-3.las')
```

```python
df = las.df()
```

```python
df.reset_index(inplace=True)
df.rename(columns={'DEPT':'DEPTH'}, inplace=True)
df.head()
```

## Plotting the Data
### Simple Plot Fills

```python
# setup figure and plot
#setup axes limits
plt.figure(figsize=(5,8))
plt.plot(df['GR'], df['DEPTH'], c='black', lw=0.5)

```

```python
# Add fills_between
# setup figure and plot
#setup axes limits
plt.figure(figsize=(5,8))
plt.plot(df['GR'], df['DEPTH'], c='black', lw=0.5)


plt.xlim(0,150)
plt.ylim(2500,200)
```

```python
# Add fills_between
# setup figure and plot
#setup axes limits
plt.figure(figsize=(5,8))
plt.plot(df['GR'], df['DEPTH'], c='black', lw=0.5)
#adding fill
plt.fill_betweenx(df['DEPTH'], 0,df['GR'], facecolor='red')
plt.xlim(0,150)
plt.ylim(2500,2000)
```

```python
# Add fills_between
# setup figure and plot
#setup axes limits
plt.figure(figsize=(5,8))
plt.plot(df['GR'], df['DEPTH'], c='black', lw=0.5)
#adding fill to left
plt.fill_betweenx(df['DEPTH'], 0,df['GR'], facecolor='red')
#adding fill to right
plt.fill_betweenx(df['DEPTH'], df['GR'],150, facecolor='yellow')


plt.xlim(0,150)
plt.ylim(2500,2000)
```

Fill to a fixed Value

```python
# Add fills_between
# setup figure and plot
#setup axes limits
plt.figure(figsize=(5,8))
plt.plot(df['GR'], df['DEPTH'], c='black', lw=0.5)
#using where argument to fill to a fixed value
plt.fill_betweenx(df['DEPTH'], 0,df['GR'],where=df['GR']<=50, facecolor='red')
plt.fill_betweenx(df['DEPTH'], df['GR'],50, where=df['GR']>=50, facecolor='yellow')


plt.xlim(0,150)
plt.ylim(2500,2000)
```

## Color Map Fills

### Variable Fill to left


```python
#setup plot values and color index
left_value=0
right_value=150
span=abs(left_value-right_value)

cmap=plt.get_cmap('hot_r')

color_index=np.arange(left_value,right_value,span/100)
```

```python
plt.figure(figsize=(5,8))
plt.plot(df['GR'], df['DEPTH'], c='black', lw=0.5)

plt.xlim(0,150)
plt.ylim(2500,2000)
plt.xlabel("Gamma Ray - API",fontsize=14)
plt.ylabel("Depth - m",fontsize=14)
for index in sorted(color_index):
    index_value= (index-left_value)/span
    color = cmap(index_value)
    print(f'{index_value}\t{color}')
```

```python
plt.figure(figsize=(5,8))
plt.plot(df['GR'], df['DEPTH'], c='black', lw=0.5)

plt.xlim(0,150)
plt.ylim(2500,2000)
plt.xlabel("Gamma Ray - API",fontsize=14)
plt.ylabel("Depth - m",fontsize=14)
for index in sorted(color_index):
    index_value= (index-left_value)/span
    color = cmap(index_value)
    # print(f'{index_value}\t{color}')
    plt.fill_betweenx(df['DEPTH'],left_value,df['GR'], where=df['GR']>=index, color=color)
```

Variable Fill to the right

```python
plt.figure(figsize=(5,8))
plt.plot(df['GR'], df['DEPTH'], c='black', lw=0.5)

plt.xlim(0,150)
plt.ylim(2500,2000)
plt.xlabel("Gamma Ray - API",fontsize=14)
plt.ylabel("Depth - m",fontsize=14)
for index in sorted(color_index):
    index_value= (index-left_value)/span
    color = cmap(index_value)
    # print(f'{index_value}\t{color}')
    plt.fill_betweenx(df['DEPTH'],right_value,df['GR'], where=df['GR']>=index, color=color)
```
