# Core Data Visualization using Matplotlib

Core Data Visualization is very important in understanding the core and getting the most useful information out of it. We can plot multiple subplots using matplotlib modules within python. Python provides a lot of functionality when working with any kind of petrophysical data.

## importing Modules

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../Data/inputs/15_9-19A-CORE.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

Here in Dataset we can see different columns;
- **DEPTH** column with drillers point of view
- **OrigDepth** with Original Depth
- **Core_NO** and **SAMPLE** column with informations about identity of core and samples
- **CKHG, CKHL, CKVG, CKVL** are a variety of Permeability measurements
- **CPOR** and **CPORV** are porosity measurements
- **So** for oil saturation values
- **Sw** for Water Saturatoin values
- **CGD** and **CGDV** are measure Grain Densities


## Visualizations using Subplots

```python
# Create the figure
fig, ax = plt.subplots(figsize=(10,10))

# creating axes
ax1 = plt.subplot2grid(shape=(3,3), loc=(0,0), rowspan=3)
ax2 = plt.subplot2grid(shape=(3,3), loc=(0,1), rowspan=3)
ax3 = plt.subplot2grid(shape=(3,3), loc=(0,2))
ax4 = plt.subplot2grid(shape=(3,3), loc=(1,2))
ax5 = plt.subplot2grid(shape=(3,3), loc=(2,2))

# adding data plot ax1
ax1.scatter(df['CPOR'], df['DEPTH'], c='red', marker='.')
ax1.set_xlim(0,50)
ax1.set_ylim(4010,3825)
ax1.set_title('Core Porosity')
ax1.grid()

#adding data to ax2

ax2.scatter(df['CKHG'], df['DEPTH'], c='green', marker='.')
ax2.set_xlim(0.01,10000)
ax2.semilogx()
ax2.set_ylim(4010,3825)
ax2.set_title('Core Permiability')
ax2.grid()

#adding data to ax3

ax3.scatter(df['CPOR'], df['CKHG'], alpha=0.5, marker='.')
ax3.set_xlim(0,50)
ax3.semilogy()
ax3.set_title('Poro-Perm Scatter Plot')
ax3.set_xlabel('Core Porosity (%)')
ax3.set_ylabel("Core Permeability (mD) ")
ax3.grid()


# adding histogram in ax4
ax4.hist(df['CPOR'], bins=30, edgecolor='black', color='red', alpha=0.6)
ax4.set_xlabel('Core Porosity (%)')

# adding histogram in ax5
ax5.hist(df['CGD'], bins=30, edgecolor='black', color='purple', alpha=0.6)
ax5.set_xlabel('Core Grain Density (%)')



plt.tight_layout()
plt.show()





```
