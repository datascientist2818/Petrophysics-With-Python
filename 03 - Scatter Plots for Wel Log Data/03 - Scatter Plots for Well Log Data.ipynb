# Creating Scatterplots (Crossplots) of Well Log Data

## Introduction
Scatterplots are commonly used data visualisation tool to allow us to identify and determine if there is a relationship between two variables. We will also be able to tell if that relationship is a strong one or if there is no relationship.

Within Petrophysics, Scatterplots or Crossplots, are routinely used as part of the interpretation workflow. They allow us to determine key interpretation parameters such as
- clay and shale end points for our clay or shale volume calculations
- outlier detection
- Lithology identificatoin
- hydrocarbon identification
- rock typing
- regression analysis and many more..!

## Importing Modules and Loading LAS Files

Here, We will import modules including, LASIO, MATPLOTLIB and pandas to work with log files.

```python
import pandas as pd
import lasio
import matplotlib.pyplot as plt
```

```python
las = lasio.read("15-9-19_SR_COMP.LAS")
```

```python
# converting to pandas dataframe object
df = las.df()
```

Getting information, Summary statistics for the data using .describe() and .info() method

```python
df.describe()
```

```python
df.info()
```

```python
df.head()
```

## Crossplots / Scatterplots
Now, Crossplots can be created, we will use density v/s neutron porosity measurements. These two measurements are often plotted like this and can tell us a number of different things about the intervals logged, including hydrocarbon presence, lithology, and bad data etc.


```python
# Set up the scatter plot
plt.scatter(x='NEU', y='DEN', data=df )
plt.show()
```

We can set default plot size for our scatterplots using plt.rcParams

```python
plt.rcParams['figure.figsize'] = (8,8)
```

```python
# set up scatterplot
plt.scatter(x='NEU', y='DEN', data =df)
#change X, Y ranges
plt.xlim(-5,60)

#for the y axis, we need to flip by passing in the scale values in reverse order
plt.ylim(3.0,1.5)

plt.show()
```

## Adding Labels to the Axes

The scatterplot above is not much use to anyone else, there are no labels or units on the axes. So we need to tell readers of the plot what is plotted against what.

We can add these using plt.xlabel and plt.ylabel

```python
plt.rcParams['figure.figsize'] = (8,8)
```

```python
# set up scatterplot
plt.scatter(x='NEU', y='DEN', data =df)
#change X, Y ranges
plt.xlim(-5,60)

#for the y axis, we need to flip by passing in the scale values in reverse order
plt.ylim(3.0,1.5)

# Adding labels
plt.ylabel('Bulk Density (DEN) - g/cc' , fontsize=14)
plt.xlabel('Neutron Porosity (NEU) - %' , fontsize=14)
plt.show()
```

Now we have added data points and label to make it easier for others to see and gain insights.

Now we can add another third variable using colours.

```python
# set up scatterplot
plt.scatter(x='NEU', y='DEN', c='GR',data =df, vmin =0,vmax=100)
#change X, Y ranges
plt.xlim(-5,60)

#for the y axis, we need to flip by passing in the scale values in reverse order
plt.ylim(3.0,1.5)

# Adding labels
plt.ylabel('Bulk Density (DEN) - g/cc' , fontsize=14)
plt.xlabel('Neutron Porosity (NEU) - %' , fontsize=14)

plt.show()
```

## Changing Colormap and Adding Colorbar

```python
# set up scatterplot
plt.scatter(x='NEU', y='DEN', c='GR',data =df, vmin =0,vmax=100, cmap='rainbow')
#change X, Y ranges
plt.xlim(-5,60)

#for the y axis, we need to flip by passing in the scale values in reverse order
plt.ylim(3.0,1.5)

# Adding labels
plt.ylabel('Bulk Density (DEN) - g/cc' , fontsize=14)
plt.xlabel('Neutron Porosity (NEU) - %' , fontsize=14)
# Make the colorbar show
plt.colorbar(label='Gamma Ray - API')
plt.show()
```

## Adding Gridlines & Plot Styling

Style sheets allows us to control the look and feel of the plots.

```python

#Set style sheet to plot
plt.style.use('bmh')
# set up scatterplot
plt.scatter(x='NEU', y='DEN', c='GR',data =df, vmin =0,vmax=100, cmap='rainbow')
#change X, Y ranges
plt.xlim(-5,60)

#for the y axis, we need to flip by passing in the scale values in reverse order
plt.ylim(3.0,1.5)

# Adding labels
# Adding labels
plt.ylabel('Bulk Density (DEN) - g/cc' , fontsize=14)
plt.xlabel('Neutron Porosity (NEU) - %' , fontsize=14)
# Make the colorbar show
plt.colorbar(label='Gamma Ray - API')
plt.show()
```

## Changing Data

If we wanted to view other curves on our plot, we can swap our variables in the plt.scatter line. Like we can switch NEU data for AC (Acoustic Compressional Slowness). Once we have done this we can quickly update the scales and the labels.


```python

#Set style sheet to plot
plt.style.use('bmh')
# set up scatterplot
plt.scatter(x='AC', y='DEN', c='GR',data =df, vmin =0,vmax=100, cmap='rainbow')
#change X, Y ranges
plt.xlim(40,240)

#for the y axis, we need to flip by passing in the scale values in reverse order
plt.ylim(3.0,1.5)

# Adding labels
plt.ylabel('Bulk Density (DEN) - g/cc' , fontsize=14)
plt.xlabel('Acoustic Compressional (AC) - us/ft' , fontsize=14)
# Make the colorbar show
plt.colorbar(label='Gamma Ray - API')
plt.show()
```
